```python
%matplotlib inline
import cv2
import numpy
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 30
print(cv2.__version__)
```

```python
image = cv2.imread("images/testsubtitle_5.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # cv use bgr but plt use rgb
plt.imshow(image)
plt.show()
```

```python
black_mask = cv2.inRange(image, (0,0,0), (10,10,10))
plt.imshow(black_mask,cmap = plt.cm.gray)
plt.show()
```
