```python
#เตรียมการ library
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10,20
print(cv2.__version__)
```

```python
#ขั้น 0: โหลดรูปภาพต้นฉบับ
ori_image = cv2.imread("original.jpg")
ori_image = cv2.cvtColor(ori_image, cv2.COLOR_BGR2RGB) # cv2 use bgr but plt use rgb
plt.imshow(ori_image)
```

```python
#ขั้น 2 สร้าง mask ของลายน้ำ (อันนี้สร้างด้วยมือ)
water_mask = cv2.imread("watermask.png")
water_mask = cv2.cvtColor(water_mask, cv2.COLOR_BGR2GRAY) # cv2 use bgr but plt use rgb
plt.imshow(water_mask,cmap = plt.cm.gray)
```

```python
#ขั้น 3 ลบลายน้ำ
out = cv2.inpaint(ori_image,water_mask,3,cv2.INPAINT_TELEA)
plt.imsave("clean.png",out)
plt.imshow(out)
```

```python
#ขั้นที่ 4: Image SuperResolution ด้วย Waifu2x
#code: https://github.com/nagadomi/waifu2x | paper: https://arxiv.org/abs/1501.00092
#เนื่อจาก Waifu2x ไม่รองรับภาษา python จึงเซฟไปทำด้านนอก
clean_image = cv2.imread("waifu2x.png")
clean_image = cv2.cvtColor(clean_image, cv2.COLOR_BGR2RGB) 
plt.imshow(clean_image)
```

```python
#ขั้น 5 สร้าง mask ของรอยแตก (สร้างด้วยมืออีกเช่นกัน)
crack_mask = cv2.imread("crackmask.png")
crack_mask = cv2.cvtColor(crack_mask, cv2.COLOR_BGR2GRAY) # cv2 use bgr but plt use rgb
plt.imshow(crack_mask,cmap = plt.cm.gray)
```

```python
#ขั้นที่ 6 เสร็จสมบูรณ์
result = cv2.inpaint(clean_image,crack_mask,5,cv2.INPAINT_TELEA)
plt.imsave("result.png",result)
plt.imshow(result)
```
