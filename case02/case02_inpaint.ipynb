```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ori_image = cv2.imread("case02.png")
```

```python
ori_image=  cv2.cvtColor(ori_image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(ori_image)
```

```python
mask_image = cv2.imread("case02_mask.png")
mask_image =  cv2.cvtColor(mask_image,cv2.COLOR_BGR2GRAY)
plt.imshow(mask_image,cmap = plt.cm.gray)
```

```python
out_image = cv2.inpaint(ori_image,mask_image,5,cv2.INPAINT_TELEA)
plt.imshow(out_image)
```

```python
plt.imsave("case02_result_2014.png",out_image)
```
