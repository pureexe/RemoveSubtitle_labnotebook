```python
%matplotlib inline
import cv2
import numpy
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 30
print(cv2.__version__)
```

```python
#load image 01 for test
image = cv2.imread("images/testsubtitle_1.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # cv use bgr but plt use rgb
plt.imshow(image)
plt.show()
```

```python
black_mask = cv2.inRange(image, (0,0,0), (10,10,10))
plt.imshow(black_mask,cmap = plt.cm.gray)
plt.show()
```

```python
white_mask = cv2.inRange(image, (245,245,245), (255,255,255))
plt.imshow(white_mask,cmap = plt.cm.gray)
plt.show()
```

```python
merge_mask = white_mask + black_mask
plt.imshow(merge_mask,cmap = plt.cm.gray)
plt.show()
```

```python
closing_mask = cv2.morphologyEx(merge_mask, cv2.MORPH_CLOSE, numpy.ones((3,3),numpy.uint8))
plt.imshow(closing_mask,cmap = plt.cm.gray)
plt.show()
```
